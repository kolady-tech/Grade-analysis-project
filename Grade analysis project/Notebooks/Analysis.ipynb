```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\kolad\OneDrive\Desktop\DataLab\Projects\Grade analysis project\data\cleaned_data.csv")
```

```python
df["G Avg"]=(df["G1"]+df["G2"]+df["G3"])/3
age_avg=df.groupby("age")["G Avg"].mean()
age_avg
```

```python
age_avg.plot(kind="bar")
plt.grid(True,alpha=0.3)
plt.title("Trend of Average G value over Age")
plt.ylabel("Average G")
plt.show()
```

```python
spread=age_avg.max() - age_avg.min()
spread
```

```python
corr = df["age"].corr(df["G Avg"])
corr
```

```python
Key Insights:
1.Highest Av G age group:20
2.Lowest Av G age group:22
3.Av G spread:The spread of average grades across age groups is approximately 6.5 marks, indicating a substantial variation in academic performance between the highest and lowest performing ages.
4.Any Av G-Age correlation:The plot shows an overall negative correlation between age and average grades, indicating that older students tend to have lower academic performance, with a minor anomaly at age 20 due to limited data.
    The Pearson correlation coefficient between age and average grade is −0.145, indicating a weak negative relationship. While grades tend to decrease slightly with increasing age, the association is not strong.
5.Any visible anomalies:The data follows downward trend except for a spike at age group 20.
    This is unusual and likely because:
    Very few students are age 20
    One or two high scorers pull the average up
    So this is a statistical anomaly / outlier effect.
```
